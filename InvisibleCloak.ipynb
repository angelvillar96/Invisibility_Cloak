### Capturing a first frame
> -  This frame is used to obtain the pixels to replace the invisible cloak

```python
#Creating a video capture object
cap = cv2.VideoCapture(0)

#Waiting some time to avoid transitories
time.sleep(3)

#capturing the image (the loop improves quality)
background=0
for i in range(30):
    ret,background = cap.read()

plt.imshow(background[:,:,::-1])
plt.title("Static Background")
plt.show()
```

### Capturing image
> -  Capturing an image to apply the algorithm and test the performance

```python
ret,img = cap.read()
hsvImage = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(img[:,:,::-1])
plt.title("Image to work with")
plt.show()
```

### Obtaining the pixels with the yellowish value high enough
> -  Converting the image to a HSV Color Space
> -  Creating a mask for the pixels with the desired hue, saturation and value (orange tones in our case)

```python
#converting to hsv color space

plt.imshow(img[:,:,::-1])
plt.title("Image Being Processed")
plt.show()

#using a mask for the desired color
#lower and upper bound contain the minimum and maximum HSV values to detect
lowerBound = np.array([15,5,5])
upperBound = np.array([60,255,255])
yellowMask = cv2.inRange( hsvImage , lowerBound , upperBound)

#deleting really small regions
mask = cv2.morphologyEx(yellowMask, cv2.MORPH_OPEN, np.ones((32,32),np.uint8))

#obtaining an image with the mask
cloakImage = np.zeros( (background.shape) )
cloakImage[mask==255,:] = img[mask==255,:]/255
plt.imshow(cloakImage[:,:,::-1])
plt.title("Mask")
plt.show()

#getting the invisible-cloak image
outputImage = np.copy( img )
outputImage[np.where(mask!=0)] = background[np.where(mask!=0)]
plt.imshow(outputImage[:,:,::-1])
plt.title("Magic!!")
plt.show()
```
